# BLAST

Gather all arrays found by CRISPRCasFinder etc. in all_arrays.fasta and all arrays that seem to be true positives in found_arrays_meta.fasta

```bash
makeblastdb -in /home/Path/to/all_arrays.fasta -dbtype nucl -out /home/Path/to/DB/all_arrays_DB
```

```bash
blastn -db /home/Path/to/DB/all_arrays_DB -query /home/Path/to/query/found_arrays_meta.fasta \
       -task blastn-short -outfmt "6 qseqid sseqid pident length qlen evalue bitscore" -evalue 1000000 -word_size 10 -dust no -soft_masking false \
       > /home/Path/to/arrays_found.tab
```

```bash
awk '{ 
    pident=$3; len=$4; qlen=$5; 
    if (len >= 69) {
        print 
    }
}' /home/Path/to/arrays_found.tab > /home/Path/to/arrays_found_filtered_set_length_69.tab
```
