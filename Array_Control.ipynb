# BLAST

Finding all true positive arrays in Crass or MCAAT output
(Step 1 is in Getting_Repeats_Spacers_and_Arrays.ipynb)

```bash
makeblastdb -in /home/Path/to/arrays/mcaat_R1_und_R2.fasta -dbtype nucl -out /home/Path/to/Array_BLAST/Erster_MCAAT/array_DB
```

```bash
blastn -db /home/Path/to/Array_BLAST/Tiefer_Crass/array_DB -query /home/Path/to/Array_BLAST/query/arrays.fasta \
    -outfmt "6 qseqid sseqid pident length qlen evalue bitscore sstart send" \
    > /home/path/to/Array_BLAST/Tiefer_Crass/arrays_found_3.tab

awk '{ 
    pident=$3; len=$4; qlen=$5; 
    # Check: Identität >= 80% (pident >= 80 &&) UND Länge des Alignments mindestens 80 % des spacers (kein 4 basen alignment bei sequenzen die 20 basen haben)
    if (len >= 69) {
        print 
    }
}' /home/Path/to/Array_BLAST/Tiefer_Crass/arrays_found_3.tab > /home/Path/to/Array_BLAST/Tiefer_Crass/arrays_found_69_3.tab
```
