# Crass
## Repeats and Spacers

```bash
awk '
function out(){
if(!gid) return
print drh
print dr
for(i=1;i<=n;i++){
print sph[i]
print sp[i]
}
}

/<group .*drseq=/{
out()
match($0,/gid="G([0-9]+)"/,g)
match($0,/drseq="([ACGT]+)"/,d)
gid=g[1]; dr=d[1]
drh=">DR|Group|" gid "|"
n=0
}

/<spacer /{
match($0,/spid="SP([0-9]+)"/,s)
match($0,/seq="([ACGT]+)"/,q)
sph[++n]=">SP|Group|" gid "|" s[1] "|"
sp[n]=q[1]
}

END{ out() }
' /home/path/to/input_dir/Crass/crass.crispr > /home/path/to/output_dir/repeats_und_spacer/crass_R1_und_R2.fasta
```

## Arrays

```bash
awk '
/^>/{
    split(substr($0,2),a,"|")
    t=a[1]; g=a[3]
    if(t=="SP") i[g]++
    next
}

t=="DR"{ dr[g]=$0 }
t=="SP"{ sp[g,i[g]]=$0 }

END{
    for(g in dr){
        printf(">array|Group|%s|\n%s",g,dr[g])
        for(j=1;j<=i[g];j++) printf("%s%s",sp[g,j],dr[g])
        print ""
    }
}' /home/path/to/input_dir/repeats_und_spacer/crass_R1_und_R2.fasta > /home/path/to/output_dir/arrays/crass_R1_und_R2.fasta
```

# MCAAT
## Repeats and Spacers

```bash
awk '
function out(){
    if(!r) return
    print ">DR|Group|"g"|"; print r
    for(i=1;i<=n;i++){ print ">SP|Group|"g"|"i"|"; print s[i] }
}

/^-+/ { d++; if(d==2)m=1; else if(d==3)m=2; next }
/^Number of Spacers/ { out(); delete s; n=0; r=""; d=0; next }
/^$/ { next }

m==1 { g++; r=$0 }
m==2 { s[++n]=$0 }

END{ out() }
' /home/path/to/input_dir/MCAAT/CRISPR_Arrays.txt > /home/path/to/output_dir/repeats_und_spacer/mcaat_R1_und_R2.fasta
```

## Repeats and Spacer

```bash
#The name of the Output .txt is chosen by you when using MinCED  
#I chose a similar patten to my other files
```

```bash
awk '
function out(){
    if(!r) return
    print dh; print r
    for(i=1;i<=n;i++){ print sh[i]; print ss[i] }
}

FNR==1 && NR>1{ delete seen }

/^[0-9]+/{
    rep=$2; sp=$3

    if(!seen[rep]++){
        out()
        r=rep
        dh=">DR|Group|"++g"|"
        n=0
        delete sh; delete ss
    }

    if(sp){
        ss[++n]=sp
        sh[n]=">SP|Group|"g"|"n"|"
    }
}

END{ out() }
' /home/path/to/input_dir/MinCED/minced_R1.txt > /home/path/to/output_dir/repeats_und_spacer/minced_R1_und_R2.fasta

awk '
function out(){
    if(!r) return
    print dh; print r
    for(i=1;i<=n;i++){ print sh[i]; print ss[i] }
}

FNR==1 && NR>1{ delete seen }

/^[0-9]+/{
    rep=$2; sp=$3

    if(!seen[rep]++){
        out()
        r=rep
        dh=">DR|Group|"++g"|"
        n=0
        delete sh; delete ss
    }

    if(sp){
        ss[++n]=sp
        sh[n]=">SP|Group|"g"|"n"|"
    }
}

END{ out() }
' /home/path/to/input_dir/MinCED/minced_R2.txt >> /home/path/to/output_dir/repeats_und_spacer/minced_R1_und_R2.fasta
```
